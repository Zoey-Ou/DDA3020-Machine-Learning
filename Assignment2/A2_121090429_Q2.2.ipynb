```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
# import data
current_directory = os.getcwd()

y_train=pd.read_csv(os.path.join(current_directory,'y_train.csv'))['target']
y_test=pd.read_csv(os.path.join(current_directory,'y_test.csv'))['target']

class_labels = ['setosa','versicolor','virginica']
label_values = np.unique(y_train)

x_train=pd.read_csv(os.path.join(current_directory,'x_train.csv'))
x_test=pd.read_csv(os.path.join(current_directory,'x_test.csv'))
```

### Q2.2.1 Calculation using Standard SVM Model

```python
print('Q2.2.1 Calculation using Standard SVM Model: \n')
for class_label, label_value in zip(class_labels, label_values):
    # Convert labels to binary (-1 or 1) for the current class
    y_train_binary = y_train.apply(lambda x: 1 if x == label_value else -1)
    y_test_binary = y_test.apply(lambda x: 1 if x == label_value else -1)

    # Create an SVM model with a linear kernel and C = 1e5
    svm_model = svm.SVC(kernel='linear', C=1e5)

    # Train the SVM model
    svm_model.fit(x_train, y_train_binary)

    # Predict on the training and testing datasets
    y_train_pred = svm_model.predict(x_train)
    y_test_pred = svm_model.predict(x_test)

    # Calculate classification errors
    train_error = np.mean(y_train_binary != y_train_pred)
    test_error = np.mean(y_test_binary != y_test_pred)

    # Output results in the specified format
    print(f'{class_label} training error: {train_error:.2f}, testing error: {test_error:.2f},')
    print(f'w_of_{class_label}: {svm_model.coef_}, b_of_{class_label}: {svm_model.intercept_},')
    print(f'support_vector_indices_of_{class_label}: {svm_model.support_}. \n')
```

### Q2.2.2 Calculate using SVM with Slack Variables (Linear Kernel)

```python
print('Q2.2.2 Calculate using SVM with Slack Variables (C = 0.2 × t, where t = 1, 2, ..., 5): \n')
t_set = [1,2,3,4,5]

for t in t_set:
    C_value = 0.2 * t
    print(f"C: {C_value},")
    for class_label, label_value in zip(class_labels, label_values):
        # Convert labels to binary (-1 or 1) for the current class
        y_train_binary = y_train.apply(lambda x: 1 if x == label_value else -1)
        y_test_binary = y_test.apply(lambda x: 1 if x == label_value else -1)

        # Create an SVM model with a linear kernel
        svm_model = svm.SVC(kernel='linear', C=1e5)

        # Train the SVM model
        svm_model.fit(x_train, y_train_binary)

        # Predict on the training and testing datasets
        y_train_pred = svm_model.predict(x_train)
        y_test_pred = svm_model.predict(x_test)

        # Calculate classification errors
        train_error = np.mean(y_train_binary != y_train_pred)
        test_error = np.mean(y_test_binary != y_test_pred)
        support_indices = list(svm_model.support_)
        slack_variables = np.maximum(0, 1 - y_train_binary.loc[support_indices] * svm_model.decision_function(x_train.loc[support_indices]))

        # Output results in the specified format
        print(f'{class_label} training error: {train_error:.2f}, testing error: {test_error:.2f},')
        print(f'w_of_{class_label}: {svm_model.coef_}, b_of_{class_label}: {svm_model.intercept_},')
        print(f'support_vector_indices_of_{class_label}: {svm_model.support_},')
        print(f'slack_variable_of_{class_label}: {slack_variables}.')
        
    print("-----------------------------------------\n")
```
